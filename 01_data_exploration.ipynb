```python
# Cell 1: بارگذاری کتابخانه‌ها
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
# Cell 2: بارگذاری داده
df = pd.read_csv('../data/emotions.csv')
print("Shape:", df.shape)
display(df.head())
```

```python
# Cell 3: بررسی اطلاعات اولیه
print("Data Info:")
df.info()
```

```python
# Cell 4: آمار توصیفی
print("Descriptive Statistics:")
display(df.describe())
```

```python
# Cell 5: بررسی مقادیر null
print("Missing Values:")
missing = pd.DataFrame({
    'missing_count': df.isnull().sum(),
    'missing_percentage': df.isnull().mean() * 100
})
display(missing[missing['missing_count'] > 0])
```

```python
# Cell 6: بررسی توزیع کلاس‌ها
plt.figure(figsize=(8, 5))
df['label'].value_counts().plot(kind='bar')
plt.title('Distribution of Emotion Labels')
plt.show()
```

```python
# Cell 7: بررسی همبستگی (برای تعداد ویژگی‌های محدود)
if df.shape[1] < 20:  # اگر تعداد ویژگی کم است
    plt.figure(figsize=(12, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()
```
